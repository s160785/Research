```python
import pandas as pd
import numpy as np
```

### Comparing all the mixes with other mixes in 28 days Water Curing

```python
df_28 = pd.DataFrame({
    "Mix_ID" : ['MO','M1','M2','M3','M4','M5','M6','M7','M8','M9'],
    "Fiber" : ["None","SF","SF","SF","DPF","DPF","DPF","PPF","PPF","PPF"],
    "Vf" : [0,0.2,0.6,1,0.2,0.6,1,0.2,0.6,1],
    "Mpa": [64.5,64.1,63.8,63.2,58.1,55.0,69.2,70.1,66.6,55.6],
    "SD": [0.4369786417,4.980013387,2.291164187,2.759226945,0.9504384953,0.4582575695,4.221958471,2.20290717,3.245620126,1.414213562]
})
df_28
```

```python
df_28_all_tests = df_28
df_28_all_tests
```

```python
df_28_all_tests.set_index("Mix_ID",inplace=True)
```

```python
df_28_all_tests = df_28_all_tests
df_28_all_tests
```

```python
mpa = list(df_28["Mpa"])
mpa
df_28_all_tests
```

```python
for k,i in enumerate(mpa):
    l = []
    for j in mpa:
        l.append(float((float(j-i)*100)/i))
    
    row = f"M{k}"
    print(len(l),row)
    df_28_all_tests[row] = l 
```

```python
df_28_all_tests
```

```python
df_28_all_tests["Total"] = df_28_all_tests.drop(["Mpa","Vf","Fiber"],axis = 1).apply(np.sum,axis =1)
```

```python
df_28_all_tests
```
